```python
from typing import List, Dict, Any

def merge_lists_by_id(*lists: List[Dict[str, Any]], key: str = "student_id") -> List[Dict[str, Any]]:
    if not lists:
        return []

    # 1. строим индексы для всех списков, кроме первого
    indices = []
    for lst in lists[1:]:
        idx = {d[key]: d for d in lst}
        indices.append(idx)

    base = lists[0]
    result = []

    for item in base:
        k = item[key]
        merged = item.copy()
        for idx in indices:
            if k in idx:
                # значения из поздних словарей перезапишут поля при конфликте
                merged |= idx[k]          # Python 3.9+ [web:7][web:10]
        result.append(merged)

    return result
```

### **Посчитайте Retention 15 дня (в процентах) для пользователей, зарегистрированных в январе**

Cохраните результат в переменную `retention_15_day`

**Примечание:** результат округлите до 5 знаков после запятой

```python
# Ваше решение
# Ищем в registrations, сколько зарегистрировалось в январе

number_retention_15_day = 0
s1 = []

from datetime import datetime
fmt = '%Y-%m-%d'  # день.месяц.год

for i in registrations:
  if '-01-' in i['registration_date']:
    s1.append({'user_id': i['user_id'], 'registration_date' : i['registration_date']})
    for j in entries:
      if i['user_id'] == j['user_id'] and (datetime.strptime(j['entry_date'], fmt)  - datetime.strptime(i['registration_date'], fmt)).days == 15:
        number_retention_15_day += 1
        break



retention_15_day = round(number_retention_15_day/len(s1)*100, 5)
retention_15_day
```

### **Посчитайте Mau, Wau, Dau усредненные**

Сохраните результат в переменные `avg_mau`, `avg_wau`, `avg_dau` соответственно

**Примечание:** результаты округлите до 5 знаков после запятой

```python
# Ваше решение
all_dates = [datetime.strptime(e['entry_date'], '%Y-%m-%d') for e in entries]
min_date = min(all_dates)
max_date = max(all_dates)

# MAU - усредненный за все месяцы
mau_values = []
current_date = datetime(min_date.year, min_date.month, 1)
while current_date <= max_date:
    next_month = (current_date.month % 12) + 1
    next_year = current_date.year if current_date.month < 12 else current_date.year + 1
    month_start = current_date.strftime('%Y-%m')

    users_in_month = set()
    for e in entries:
        if e['entry_date'].startswith(month_start):
            users_in_month.add(e['user_id'])

    if len(users_in_month) > 0:
        mau_values.append(len(users_in_month))

    current_date = datetime(next_year, next_month, 1)

avg_mau = round(sum(mau_values) / len(mau_values), 5) if mau_values else 0
avg_mau
```

```python
from collections import defaultdict
from datetime import datetime
import csv

with open('entries.csv', mode='r', encoding='utf-8') as f:
    entries = list(csv.DictReader(f, delimiter=';'))

# WAU: группируем по ISO-неделям
weeks = defaultdict(set)
for e in entries:
    dt = datetime.strptime(e['entry_date'], '%Y-%m-%d')
    week_key = dt.isocalendar()[:2]  # (год, номер_недели_ISO)
    weeks[week_key].add(e['user_id'])

wau_values = [len(v) for v in weeks.values()]
avg_wau = round(sum(wau_values) / len(wau_values), 5)
```
