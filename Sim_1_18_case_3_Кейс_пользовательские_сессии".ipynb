```python
import pandas as pd

df = pd.read_csv('sessions.csv')
df
```

```python
import csv
import json
from datetime import datetime

# Функция для преобразования дат в datetime
def parse_dates(obj):
    """Рекурсивно обходит структуру и конвертирует строки формата datetime"""
    if isinstance(obj, list):
        return [parse_dates(item) for item in obj]
    elif isinstance(obj, str):
        try:
            return datetime.fromisoformat(obj)
        except:
            return obj
    else:
        return obj

# Читаем и парсим данные
sessions_data = []

with open('sessions.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)  # пропускаем заголовок

    for row in reader:
        user_id = row[0]
        session_json = row[1]

        # Десериализуем JSON
        session = json.loads(session_json)

        # Преобразуем даты
        session_parsed = parse_dates(session)

        sessions_data.append({
            'user_id': user_id,
            'session_start': session_parsed[0],
            'duration': session_parsed[1],
            'pages': session_parsed[2]
        })
```

#### **Задача 1**

Ваша задача написать функцию `get_time_to_conversion`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает среднее время между заходом на сайт и целевым действием.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res`

**Решение**

Напишите свое решение ниже

```python
import csv
import json
from datetime import datetime
from statistics import mean

def get_time_to_conversion(sessions):
    times_to_conversion = []

    for session in sessions:
        session_start = session['session_start']  # уже datetime

        for page in session['pages']:
            if 'conversion' in page[0]:
                conversion_time = page[1]  # уже datetime
                time_diff = (conversion_time - session_start).total_seconds()
                times_to_conversion.append(time_diff)
                break

    if times_to_conversion:
        return round(mean(times_to_conversion), 2)
    else:
        return 0.0

res = get_time_to_conversion(sessions_data)
print(res)
```

#### **Задача 2**

Ваша задача написать функцию `get_session_duration`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает сколько вообще в среднем длится сессия.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res2`

**Решение**

```python

def get_session_duration(sessions):
    """
    Подсчитывает среднюю длительность сессии.

    Args:
        sessions: список сессий с полем 'duration'

    Returns:
        float: средняя длительность в секундах, округленная до 2 знаков
    """
    durations = [session['duration'] for session in sessions]
    return round(mean(durations), 2)

res2 = get_session_duration(sessions_data)
print(res2)
```

#### **Задача 3**

Напишите функцию `get_daily_session_count`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает cколько разных заходов в течение дня делает пользователь.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res3`

**Решение**

Напишите свое решение ниже

```python
from collections import defaultdict
from statistics import mean

def get_daily_session_count(sessions):
    """
    Подсчитывает среднее количество сессий в день.

    Args:
        sessions: список сессий с 'user_id' и 'session_start'

    Returns:
        float: среднее количество сессий в день, округленное до 2 знаков
    """
    # Группируем сессии по (user_id, date)
    user_daily_sessions = defaultdict(lambda: defaultdict(int))

    for session in sessions:
        user_id = session['user_id']
        date = session['session_start'].date()
        user_daily_sessions[user_id][date] += 1

    # Собираем все количества сессий по всем дням
    all_daily_counts = []
    for user_id, daily_counts in user_daily_sessions.items():
        all_daily_counts.extend(daily_counts.values())

    # Прямое среднее по дням
    return round(mean(all_daily_counts), 2)

res3 = get_daily_session_count(sessions_data)
print(res3)
```

#### **Задача 4**

Напишите функцию `get_average_session_count`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает сколько заходов вообще в среднем делает пользователь.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res4`

**Решение**

Напишите свое решение ниже

```python

from collections import defaultdict
from statistics import mean

def get_average_session_count(sessions):
    """
    Подсчитывает среднее количество сессий на пользователя.

    Args:
        sessions: список сессий с 'user_id'

    Returns:
        float: среднее количество сессий на пользователя, округленное до 2 знаков
    """
    # Считаем сессии по пользователям
    user_session_count = defaultdict(int)

    for session in sessions:
        user_session_count[session['user_id']] += 1

    # Среднее количество сессий на пользователя
    return round(mean(user_session_count.values()), 2)

res4 = get_average_session_count(sessions_data)
print(res4)
```
