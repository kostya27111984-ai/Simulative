Напишите функцию most_frequent_weekday, которая принимает список дат в формате год-месяц-день и возвращает самый часто встречающийся день недели в этом списке (например, "Monday").

Если несколько дней встречаются одинаково часто, функция должна вернуть все эти дни.

```python
'''Фукнкция принимает список дат в формате год-месяц-день и возвращает самый часто встречающийся день недели в этом списке (например, "Monday").
Если несколько дней встречаются одинаково часто, функция должна вернуть все эти дни.'''

from datetime import datetime
from collections import Counter

def most_frequent_weekday(str_dates):
  weekdays = [datetime.strptime(i, '%Y-%m-%d').strftime('%A') for i in str_dates]

  number_of_wd = Counter(weekdays).most_common()

  high_number = number_of_wd[0][1]
  s = []
  s.append(number_of_wd[0][0])
  flag = False
  if len(number_of_wd) > 1:
    for i in number_of_wd[1:]:
      if i[1] < high_number:
        flag = True
      else:
        s.append(i[0])
      if flag:
        break
  return s
```

```python
#Оптимизированный код
from collections import Counter
from datetime import datetime

def most_frequent_weekday(dates):
    day_week = [datetime.strptime(day, "%Y-%m-%d").strftime('%A') for day in dates]
    counter = Counter(day_week)
    max_value = max(counter.values())
    return [item for item, count in counter.items() if count == max_value]
```

Python ДЗ 12: Средняя разница времени

При проведении разного рода исследований нам может потребоваться проанализировать разницу во времени между последовательными датами. Такой анализ может помочь нам понять временные промежутки между событиями и принимать решения на основе этого.

Напишите функцию, которая принимает список строк в качестве входных данных и возвращает среднюю разницу во времени между последовательными датами в секундах.

```python
from datetime import datetime, timedelta

def avg_time_diff(dates):
    dt = [datetime.strptime(day, "%Y-%m-%d %H:%M:%S") for day in dates]
    if len(dt) < 2:
      res = 0.0
    else:
      delta = [(i - j).total_seconds() for i in dt[1:] for j in dt[:-1]]
      res = sum(delta)/len(delta)
    return res
```

Python ДЗ 12: Количество измерений в интервале

Вы работаете над системой, которая анализирует данные с датчиков, которые собирают измерения через регулярные промежутки времени. Временные метки этих измерений сохраняются в виде даты и времени в списке, отсортированном в порядке возрастания. Вам нужно определить, сколько раз измерения выполняются в течение определенного интервала времени.

Напишите функцию, которая принимает на вход список дат и времени, отсортированных в порядке возрастания, и временной интервал (в минутах) и возвращает количество раз, когда даты и время в списке встречаются в пределах интервала друг от друга.

Например, если входным списком является ['2022-01-01 09:00:00', '2022-01-01 11:00:00', '2022-01-02 09:00:00'] и интервал составляет 120 минут, функция должна возвращать 1. Поскольку только разница между первыми двумя датами - лежит в пределах не больше данного нам интервала и за 120 минут проводится только одно измерение.

```python
from datetime import datetime, timedelta

def count_measurements_within_interval(dates_w_time, max_delta):
    dt = [datetime.strptime(day, "%Y-%m-%d %H:%M:%S") for day in dates_w_time]
    count = 0

    for i in range(len(dt) - 1):
        delta_minutes = (dt[i + 1] - dt[i]).total_seconds() / 60
        if delta_minutes <= max_delta:
            count += 1

    return count
```

```python
count_measurements_within_interval(['2022-01-01 09:00:00', '2022-01-01 11:00:00', '2022-01-02 09:00:00'], 120)

# 1
```

```python
count_measurements_within_interval(['2022-01-01 09:00:00', '2022-01-01 09:15:00', '2022-01-01 09:45:00'], 30)

# 2
# В данном случае первые два измерения лежат в пределах менее 30 минут - поэтому произведено 2 измерени
```

```python
count_measurements_within_interval(['2022-01-01 09:00:00', '2022-01-01 09:20:00', '2022-01-01 10:00:00'], 15)

# 0
# Ни одно из измерений не лежит в пределах 15 минут друг от друга. Соответственно,
# за 15 минут не производится ни одного измерения
```

Python ДЗ 12: Время до праздника

Напишите функцию days_to_nearest_holiday, которая принимает дату в формате год-месяц-день и возвращает количество дней до ближайшего предстоящего праздника.

Список праздников:

"Рождество": "01-07",
"День влюбленных": "02-14",
"Женский день": "03-08",
"День труда": "05-01",
"День знаний": "09-01",
"Новый год": "12-31"

```python
from datetime import date, datetime, timedelta

def days_to_nearest_holiday(str_date):
  holidays = {
    "Рождество": "01-07",
    "День влюбленных": "02-14",
    "Женский день": "03-08",
    "День труда": "05-01",
    "День знаний": "09-01",
    "Новый год": "12-31"
            }
  cur_date = datetime.strptime(str_date, "%Y-%m-%d")
  cur_year = cur_date.year
  s = []
  for _ in range(2):
      for k,v in holidays.items():
          hol_date = datetime.strptime(f'{cur_year}-{v}', "%Y-%m-%d")
          delta = (hol_date - cur_date).total_seconds()/60/60/24
          if delta >=0:
              s.append(delta)
      if len(s) > 0:
        break
      else:
        cur_year ++ 1

  return int(min(s))
```

```python
days_to_nearest_holiday('2023-01-01')

# 6
```

```python
days_to_nearest_holiday('2023-12-31')

# 0
```

Python ДЗ 12: Количество рабочих часов

Напишите функцию get_work_hours, которая принимает две даты в формате год-месяц-день и время в формате часы:минуты:секунды для каждой из них и рассчитывает интервал между ними в целых рабочих часах, учитывая только рабочие дни (понедельник - пятница) и рабочее время (с 9 утра до 6 вечера).

```python
def get_work_hours(date1, time1, date2, time2):
  dt1 = datetime.strptime(f'{date1} {time1}', "%Y-%m-%d %H:%M:%S")
  dt2 = datetime.strptime(f'{date2} {time2}', "%Y-%m-%d %H:%M:%S")

  return (dt2 - dt1).total_seconds()/3600
```

```python
from datetime import datetime, time, timedelta

WORK_START = time(9, 0, 0)
WORK_END = time(18, 0, 0)

def is_workday(dt: datetime) -> bool:
    return dt.weekday() < 5

def clamp_to_work_interval(dt: datetime) -> datetime:
    start_dt = datetime.combine(dt.date(), WORK_START)
    end_dt = datetime.combine(dt.date(), WORK_END)

    if dt < start_dt:
        return start_dt
    if dt > end_dt:
        return end_dt
    return dt

def get_work_hours(start_date: str, start_time: str, end_date: str, end_time: str) -> float:
    start = datetime.strptime(f"{start_date} {start_time}", "%Y-%m-%d %H:%M:%S")
    end = datetime.strptime(f"{end_date} {end_time}", "%Y-%m-%d %H:%M:%S")

    if end <= start:
        return 0.0

    start = clamp_to_work_interval(start)
    end = clamp_to_work_interval(end)

    if end <= start:
        return 0.0

    total = timedelta(0)
    current_date = start.date()

    while current_date <= end.date():
        if is_workday(datetime.combine(current_date, time(0, 0, 0))):
            day_start = datetime.combine(current_date, WORK_START)
            day_end = datetime.combine(current_date, WORK_END)

            interval_start = max(start, day_start)
            interval_end = min(end, day_end)

            if interval_end > interval_start:
                total += interval_end - interval_start

        current_date += timedelta(days=1)

    return int(total.total_seconds() / 3600)
```

Python ДЗ 12: Интервалы в секундах
Нам нужно рассчитать общее количество времени, охватываемое списком временных интервалов, в секундах. Если есть перекрывающиеся интервалы, их нужно учесть только один раз. Входными данными будет список строк, представляющих временные интервалы в формате "ЧЧ:ММ-ЧЧ:ММ".

Напишите функцию calculate_duration, которая принимает список временных интервалов (например, ["9:00-10:30", "11:00-12:30", "14:00-15:00"]) и возвращает общее количество времени, пройденное интервалами, в секундах.

```python
from datetime import datetime

def calculate_duration(lst):
    intervals = []

    for item in lst:
        start_str, end_str = item.split("-")
        start = datetime.strptime(start_str, "%H:%M")
        end = datetime.strptime(end_str, "%H:%M")
        intervals.append((start, end))

    intervals.sort(key=lambda x: x[0])

    total = 0
    cur_start, cur_end = intervals[0]

    for start, end in intervals[1:]:
        if start <= cur_end:
            cur_end = max(cur_end, end)
        else:
            total += int((cur_end - cur_start).total_seconds())
            cur_start, cur_end = start, end

    total += int((cur_end - cur_start).total_seconds())
    return total
```

Ввод всегда будет в формате "ГГГГ-ММ-ДД по ГГГГ-ММ-ДД" и может содержать несколько диапазонов, разделенных запятыми. Необходимо получить список дат и времени для всех дат во всех диапазонах.

Напишите функцию date_range_to_datetimes, которая принимает строку, представляющую диапазон дат, и возвращает список дат-времени для всех дат в пределах диапазона.

Пример

```python
from datetime import datetime, timedelta

def date_range_to_datetimes(str_intervals):
  lst_str_intervals = str_intervals.split(", ")
  lst_str_intervals
  total = []
  for i in lst_str_intervals:
    str_start, str_end = i.split(' по ')
    start = datetime.strptime(str_start, '%Y-%m-%d')
    end = datetime.strptime(str_end, '%Y-%m-%d')
    current_date = start
    while current_date <= end:
      total.append(current_date)
      current_date += timedelta(days = 1)
  return total
```

Входные данные представлены списком строк в формате "ЧЧ:ММ-ЧЧ:ММ".

Если имеется более двух временных интервалов, необходимо найти максимальное перекрытие между всеми парами интервалов.

Если максимальных перекрытий несколько, верните самое позднее из них

```python
from datetime import datetime, timedelta

def find_overlap(intervals):
    fmt = "%H:%M"
    parsed = []

    for item in intervals:
        s, e = item.split("-")
        start = datetime.strptime(s, fmt)
        end = datetime.strptime(e, fmt)
        parsed.append((start, end))

    best = None          # (length, start, end)
    n = len(parsed)

    for i in range(n):
        for j in range(i + 1, n):
            s1, e1 = parsed[i]
            s2, e2 = parsed[j]

            inter_start = max(s1, s2)
            inter_end = min(e1, e2)

            if inter_start < inter_end:  # есть пересечение
                length = inter_end - inter_start
                if best is None:
                    best = (length, inter_start, inter_end)
                else:
                    best_len, best_start, best_end = best
                    if length > best_len or (length == best_len and inter_start > best_start):
                        best = (length, inter_start, inter_end)

    if best is None:
        return "Совместного интервала нет"

    _, s, e = best
    return f"{s.strftime('%H:%M')}-{e.strftime('%H:%M')}"
```
